# Data Cleaning

```python
df.isnull()
```

```python
miss=df.isnull().sum()            # to shows the count of null values
miss 
```

```python
miss_per = (df.isnull().sum()/len(df))* 100
miss_per
```

```python
sns.heatmap(df.isnull())         # using heatmap to shows the null value count
```

```python
#making a copy of our dataframe
df_copy=df.copy()
df_copy.head()
```

```python
df_copy=df_copy.dropna(how='any',subset=['director','cast'])       #droping director and cast as they have a high number of missing values
df_copy.head()
```

```python
df_copy.fillna({'country':'missing','rating':'missing','duration':'missing'},inplace=True)   #dealing with other missing values which are required in further analysis so we cant drop them
```

```python
df_copy.isnull().sum()      #checking if any more missing values
```

# EDA Questions 

### How many TV shows & Movies are in the dataset?

```python
df_copy.type.value_counts().to_frame('Value_count')
```

### What is the Correlation between the features?

```python
df_copy["date_added"]=pd.to_datetime(df_copy['date_added'])
df_copy.head(2)
```

```python
df_copy['year_added'] = df_copy["date_added"].dt.year
df_copy['month_added'] = df_copy["date_added"].dt.month
df_copy.head(2)
```

```python
df_copy.dtypes
```

```python
df_copy.corr()
```

```python
plt.subplots(figsize=(5,5))
sns.heatmap(df_copy.corr(),annot=True)
plt.show()
```

### Most watched shows on the Netflix?

```python
sns.countplot(x=df_copy['type'],orient='v')
plt.xticks(rotation=90)
```

```python
# by using Pie chart
type_show = ['Movie','TV Show']
Value_count = [5522,178]
plt.pie(Value_count,labels=type_show,autopct="%2.2f%%")
plt.legend(title='Most watched shows on the Netflix')
```
